```python
import pandas as pd
import sodapy
from sodapy import Socrata

final_df = []
offset = 0

while True:
    client = Socrata("data.cityofnewyork.us", None)
    results = client.get("bqiq-cu78",
                         limit = 1000,
                         offset = offset)
    offset = offset + 1000
    final_df.extend(results)
    if len(results) < 1000:
        break

#Convert the API call into a dataframe
results = pd.DataFrame.from_records(final_df)
results
```

```python
len(results)
```

```python
countyCrime2 = results.groupby(['county'])['complaint_precinct_code'].size()
countyCrime2.head()
```

```python
import plotly.express as px

px.bar(
    countyCrime2,
    "complaint_precinct_code",
    labels={
        "complaint_precinct_code": "",
    },
    title= "Crime Count by county")
```
